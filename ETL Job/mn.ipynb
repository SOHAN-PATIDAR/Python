Q2. Deduce the percentage of developers who knew python in each country.

```python
import sqlite3
import pandas as pd

dd = pd.read_csv("survey_results_public.csv")
dd.columns = dd.columns.str.strip()

conn = sqlite3.connect('test.db') 
dd.to_sql('survey_table', conn ,if_exists='replace')
conn.close()
```

```python
conn = sqlite3.connect('test.db') 
c = conn.cursor()
ce = conn.cursor()
countries = c.execute('''
          SELECT Country, count(*) from survey_table group by Country
          ''')

pyth = ce.execute("""
                 SELECT Country, count(*) from survey_table WHERE LanguageWorkedWith LIKE '%Python%'
 group by Country
                 """)

dic = {}

for j in pyth:
    dic[j[0]] = [j[1]]

for i in countries:
    if i[0] in dic.keys():
        dic[i[0]].append(i[1])

for key,value in dic.items():
    per = ((value[0]*100)/value[1])
    print(key,per)
```

Q4. Based on this survey, what will be the most desired programming language for the year
2020?

```python
import pandas as pd

df = pd.read_csv("survey_results_public.csv")

lang = df['LanguageDesireNextYear']
dic = {}
for i in lang:
    try :
        lst = i.split(';')
    except AttributeError as e:   # handling entries of LanguageDesireNextYear = nan
        continue
    for j in lst:
        if j not in dic:
            dic[j] = 1
        else:
            dic[j]+=1

print(max(dic,key=dic.get))
```

Q5. Generate the report for the people who code as a hobby based on their gender
and continent?
 

```python
import pandas as pd

df = pd.read_csv("survey_results_public.csv")

grp = df[['Hobbyist','Country','Gender']]
grp = grp[grp['Hobbyist']=='Yes']

merged_data = grp.merge(csv1,on=["Country"])

merged_data['Gender'].mask(merged_data["Gender"].str.contains(";", case=False, na=False) == True,  "OTHERS", inplace=True)
merged_data['Gender'].mask(merged_data["Gender"].str.contains(",", case=False, na=False) == 1,  "OTHERS", inplace=True)

res = merged_data.groupby(['Gender','Continent']).size()
res
```

Q6. Generate the report for job and career satisfaction of developers based on their
gender and continent?

```python
import pandas as pd

df = pd.read_csv("survey_results_public.csv")

# satisfied_dev = res_public.copy(deep=True)
rep = df.merge(csv1, how ="left", on=["Country"])

# Replacing a new Category 'OTHERS' for ambiguous values
rep['Gender'].mask(rep["Gender"].str.contains(";", case=False, na=False) == True,  "OTHERS", inplace=True)
rep['Gender'].mask(rep["Gender"].str.contains(",", case=False, na=False) == 1,  "OTHERS", inplace=True)

# Creating a new column for each satisfaction level
rep["Job Very satisfied"] = rep['JobSat'] == "Very satisfied"
rep["Job Slightly satisfied"] = rep['JobSat'] == "Slightly satisfied"
rep["Job Slightly dissatisfied"] = rep['JobSat'] == "Slightly dissatisfied"
rep["Job Neither satisfied nor dissatisfied"] = rep['JobSat'] == "Neither satisfied nor dissatisfied"
rep["Job Very dissatisfied"] = rep['JobSat'] == "Very dissatisfied"

# Storing all data grouped by "Gender" & "Continent" in a new variable 
status_info = rep.groupby(["Gender", "Continent"]).sum()

status_info
```
